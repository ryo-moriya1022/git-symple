# **課題１**
　次の行列の積を計算せよ。
$$
\begin{pmatrix}
1 & 2 & 3 \\\ 
4 & 5 & 6
\end{pmatrix}
\begin{pmatrix}
3 & 4 \\\ 
6 & 8 \\\ 
10 & 11
\end{pmatrix}
$$

```python
import numpy as np
data1=np.array([[1,2,3],[4,5,6]])
data2=np.array([[3,4],[6,8],[10,11]])
print(np.matmul(data1, data2))
```

# **課題２**
　成績のデータを分析することを考える。ここでは、成績を模したデータから基本的な特徴を表す値を計算して取得するプログラムを作る。成績を模したデータは、科目A（例：数学）のテストの点数とする（100点満点）。人数は10人とする。点数は最初の人から順に、65点、92点、18点、68点、11点、69点、88点、59点、29点、60点である。
Numpyの配列を作成して、Numpyの機能を利用してテストの最低点、最高点、平均点を計算して表示しなさい。

ヒント：
Numpyには配列の最小値、最大値、平均値を計算する機能がある。

```python
import numpy as np
math=np.array([65,92,18,68,11,69,88,59,29,60])
print(np.amax(math))
print(np.amin(math))
print(np.mean(math))
```

# **課題３**
　課題２のデータを拡充して分析することを考える。ここではさらに科目B（例：化学）のテストの点数を追加する。人数と人は科目Aと同じである。最初の人から順に、60点、48点、54点、34点、56点、59点、73点、84点、36点、70点である。Numpyの配列に２つの科目のテストの点数を入れなさい。
そして、それぞれの科目ごとに最低点、最高点、平均点を計算して表示しなさい。


```python
import numpy as np
ca=np.array([60,48,54,34,56,59,73,84,36,70])
math=np.array([65,92,18,68,11,69,88,59,29,60])
seiseki=np.array([ca,math])
print(seiseki)
print("ca")
print("max=",np.amax(seiseki[0]))
print("min=",np.amin(seiseki[0]))
print("ave=",np.mean(seiseki[0]))
print("math")
print("max=",np.amax(seiseki[1]))
print("min=",np.amin(seiseki[1]))
print("ave=",np.mean(seiseki[1]))
```
